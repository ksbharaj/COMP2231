SO each email has any one of the 5 possible outcomes:
- Delivered Only
- Bounced Only
- Delivered and Opened Only
- Delivered and Spam
- Delivered, Opened and Clicked

```python
email_type_dict = {}
```

```python
for index, row in merged_df.iterrows():
  email_type = row['emails']
  if email_type not in email_type_dict:
    email_type_dict[email_type] = {'Delivered Only': 0, 'Bounced Only': 0, 'Delievered and Opened Only': 0,
                                   'Delivered and Spam': 0, 'Delivered, Opened and Clicked': 0}
  if pd.notna(row['bounced_email']):
    email_type_dict[email_type]['Bounced Only'] += 1
  if pd.notna(row['delivered_email']) & pd.isna(row['open_email']) & pd.isna(row['clicked_email']) & pd.isna(row['spam_email']) & \
     pd.isna(row['bounced_email']):
     email_type_dict[email_type]['Delivered Only'] += 1
  if pd.notna(row['delivered_email']) & pd.notna(row['open_email']) & pd.isna(row['clicked_email']) & pd.isna(row['spam_email']) & \
     pd.isna(row['bounced_email']):
     email_type_dict[email_type]['Delievered and Opened Only'] += 1
  if pd.notna(row['delivered_email']) & pd.isna(row['open_email']) & pd.isna(row['clicked_email']) & pd.notna(row['spam_email']) & \
     pd.isna(row['bounced_email']):
     email_type_dict[email_type]['Delivered and Spam'] += 1
  if pd.notna(row['delivered_email']) & pd.notna(row['open_email']) & pd.notna(row['clicked_email']) & pd.isna(row['spam_email']) & \
     pd.isna(row['bounced_email']):
     email_type_dict[email_type]['Delivered, Opened and Clicked'] += 1
```

```python
email_status_df = pd.DataFrame(email_type_dict).transpose().reset_index()
email_status_df.rename(columns={'index': 'Email Type'}, inplace=True)
```

```python
fig = px.bar(email_status_df, x='Email Type', y=['Delivered Only', 'Bounced Only', 'Delievered and Opened Only', 'Delivered and Spam',
                                                 'Delivered, Opened and Clicked'],
             title='Email Status by Email Type',
             labels={'value': 'Number of Emails'},
             height=600)
fig.show()
```

In terms of issues, "how You Can Optimize Your Writing" had the most bounced emails

## Question 4: Was the email "warm-up" process performed correctly?

```python
merged_df.sort_values(by="sent_email")
```

```python
len(merged_df.user_id.unique())
```

```python
timestamp_cols = ['sent_email', 'delivered_email', 'open_email', 'clicked_email', 'spam_email', 'bounced_email']
for col in timestamp_cols:
    merged_df[col] = pd.to_datetime(merged_df[col])
```

```python
def resample_merged_handling_nan(df, timestamp_cols, freq='M'):
    resampled_data = {}
    for col in timestamp_cols:
        temp_df = df.dropna(subset=[col])
        resampled_data[col] = temp_df.set_index(col).resample(freq).count()['user_id']
    return pd.DataFrame(resampled_data)

df_filtered = merged_df.dropna(subset=timestamp_cols, how='all')

# Resample by month, handling NaNs
df_resampled = resample_merged_handling_nan(merged_df, timestamp_cols)

# Plotting using Plotly
fig = px.line(df_resampled.reset_index(), x='index', y=timestamp_cols, title='Email Status Count Over Time (Each Month)', labels={'index': 'Time'})
fig.show()
```

```python
df_counts
```

```python
df_resampled_counts = pd.DataFrame()

# Resample each timestamp column individually and count occurrences, avoiding NaN values
for col in timestamp_cols:
    temp_df = df_filtered[[col]].dropna()
    temp_df[col] = temp_df[col].dt.to_period("M")
    resampled_count = temp_df[col].value_counts().sort_index()
    df_resampled_counts[col] = resampled_count

# Fill NaN values with zeros
df_resampled_counts.fillna(0, inplace=True)

# Calculate the Spam to Sent and Bounced to Sent ratios
df_resampled_counts['spam_to_sent_ratio'] = df_resampled_counts['spam_email'] / df_resampled_counts['sent_email']
df_resampled_counts['bounced_to_sent_ratio'] = df_resampled_counts['bounced_email'] / df_resampled_counts['sent_email']

# Plotting the ratios using Plotly
df_resampled_counts.index = df_resampled_counts.index.astype(str)

# Plotting the ratios using Plotly again
fig = px.line(df_resampled_counts.reset_index(), x='index', y=['spam_to_sent_ratio', 'bounced_to_sent_ratio'],
              title='Spam to Sent and Bounced to Sent Ratios Over Time (Each Month)', labels={'index': 'Time'})
fig.show()
```

Email Warming: The data doesn't strongly support that a successful email warming process has been conducted. A successful warming process would likely show a decreasing trend in Spam and Bounced rates, which isn't evident here.
